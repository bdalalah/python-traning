```python
class BankAccount:
    interest_rate = 0.05  # class attribute for interest rate (shared among all instances)
    
    def __init__(self, name, balance=0):
        self.name = name  # instance attribute for account holder's name
        self.balance = balance  # instance attribute for account balance
    
    @classmethod
    def set_interest_rate(cls, rate):
        cls.interest_rate = rate
    
    @classmethod
    def get_interest_rate(cls):
        return cls.interest_rate
    
    def add_money(self, amount):
        self.balance += amount
    
    def withdraw_money(self, amount):
        if amount <= self.balance:
            self.balance -= amount
        else:
            print("Insufficient balance.")
    
    def get_account_balance(self):
        return self.balance
    
    def calculate_annual_interest(self):
        return self.balance * self.interest_rate

# Example usage:
account1 = BankAccount("Alice", 1000)

# Add money to the account
account1.add_money(500)

# Withdraw money from the account
account1.withdraw_money(200)

# Print current balance
print("Current Balance for", account1.name, ":", account1.get_account_balance())

# Set a new interest rate
BankAccount.set_interest_rate(0.06)

# Calculate and print annual interest
annual_interest = account1.calculate_annual_interest()
print("Annual Interest for", account1.name, ":", annual_interest)
```

```python
class Person:
    def __init__(self, name, last_name, age):
        self.__name = name
        self.__last_name = last_name
        self.__age = age
    
    @property
    def name(self):
        return self.__name
    
    @name.setter
    def name(self, new_name):
        self.__name = new_name
    
    @property
    def last_name(self):
        return self.__last_name
    
    @last_name.setter
    def last_name(self, new_last_name):
        self.__last_name = new_last_name
    
    @property
    def age(self):
        return self.__age
    
    @age.setter
    def age(self, new_age):
        if isinstance(new_age, int) and new_age > 0:
            self.__age = new_age
        else:
            raise ValueError("Age must be a positive integer.")

# Example usage:
person = Person("Nona", "Dalalah", 28)

# Using getter methods
print("Name:", person.name)
print("Last Name:", person.last_name)
print("Age:", person.age)

# Using setter methods
person.name = "Banana"
person.last_name = "Ahmad"
person.age = 28

print("\nUpdated Information:")
print("Name:", person.name)
print("Last Name:", person.last_name)
print("Age:", person.age)
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

class Manager:
    def __init__(self, department):
        self.department = department

class Executive(Employee, Manager):
    def __init__(self, name, salary, department, company_car, bonus):
        Employee.__init__(self, name, salary)
        Manager.__init__(self, department)
        self.company_car = company_car
        self.bonus = bonus
    
    def __str__(self):
        return f"Executive: {self.name}, Department: {self.department}, Salary: {self.salary}, Bonus: {self.bonus}, Company Car: {self.company_car}"

# Example usage:
executive1 = Executive("Banoosh Ahmad", 600, "TAC", "Hwa ya 7srte :( ", 1000)
print(executive1)
```

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def set_name(self, name):
        self.name = name
        return self
    
    def set_age(self, age):
        self.age = age
        return self
    
    def get_person_info(self):
        return f"Name: {self.name}, Age: {self.age}"

# Example usage:
person = Person("Nona", 28)
person.set_name("Banosha").set_age(15)  # Method chaining to set name and age
info = person.get_person_info()
print(info)  
```
